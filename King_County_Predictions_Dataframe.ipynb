```python
def highlight_max(s, props=''):
    return np.where(s == np.nanmax(s.values), props, '')
```

```python
import pandas as pd
idx = pd.IndexSlice
th_props = [
  ('font-size', '16px'),
  ('text-align', 'center'),
  ('font-weight', '15px bold'),
  ('color', 'black'),
  ('background-color', 'white'),
  ('border','10px red'),
  ('hover','grey')
  ]

# Set CSS properties for td elements in dataframe
td_props = [
  ('font-size', '11px'),
   ('background-color','white'),
    ('color','black'),
    ('border','10px red'),
    
  ]
x = [('border','10px solid yellow')]
formatter= {('Bedrooms', 'Sqft_lot'): "$ {:,.1f}",}
# Set table styles
styles = [
  dict(selector="th", props=th_props),
  dict(selector="td", props=td_props),
  dict(selector="", props=x),
  dict(selector="td", props=formatter)
  ]
slice_ = idx[idx[:,'B'], :]

(house_info1
      .style.set_table_styles(styles)\
      .apply(highlight_max, props='color:red;', axis=1, subset=slice_)\
      .set_properties(**{'background-color': '#ffffb3'}, subset=slice_))
```
